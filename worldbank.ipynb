```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('inflation.csv', skiprows=3)
dados
```

```python
pais = dados[dados['Country Name'] == 'United States']
pais
```

```python
anos = pais['2019'], pais['2020'], pais['2021'], pais['2022']
```

## Gráfico em linha que nos o aumento (%) nos EUA. ↓↓↓

```python
x = [2019, 2020, 2021, 2022]
y = [pais['2019'], pais['2020'], pais['2021'], pais['2022']]

plt.plot(x, y, color='Blue')

plt.ylabel("Porcentagem do aumento")
plt.xlabel("Anos")
plt.title("Aumento de 2019 - 2022")
plt.xticks(x)
plt.yticks([0, 2, 4, 6, 8, 10])
plt.show()
```

```python
fig, ax = plt.subplots()

prices = [pais['2019'].values[0], pais['2020'].values[0], pais['2021'].values[0], pais['2022'].values[0]]
years = [2019, 2020, 2021, 2022]
bar_label = ['2019', '2020', '2021', '2022']
bar_color = plt.get_cmap('Blues')(np.linspace(0.5, 1.1, len(prices)))

ax.bar(years, prices, label = bar_label, color = bar_color)

ax.set_title("Aumento dos preços por conta da inflação - EUA")
ax.set_ylabel("Aumento dos preços em %")
ax.legend(title="Anos")

plt.xticks(x)
plt.yticks([0, 2, 4, 6, 8, 10])

plt.show()
```

## Histograma do aumento da inflação em todos os países no ano de 2019 a 2022. ↓↓↓

```python
dados.hist(column='2019', bins=150)
dados.hist(column='2020', bins=150)
dados.hist(column='2021', bins=150)
dados.hist(column='2022', bins=150)
plt.show()
```

## Gráfico em Scatter que nos mostra o aumento em cada ano nos EUA. ↓↓↓

```python
fig = plt.figure(figsize=(4, 3))

x = ['2019', '2020', '2021', '2022']
y = [pais['2019'].values[0], pais['2020'].values[0], pais['2021'].values[0], pais['2022'].values[0]]

total = y[0] + y[1] + y[2] + y[3]

porcentagem = [((y[0] / total) * 100) * 5, ((y[1] / total) * 100) * 5, ((y[2] / total) * 100) * 5, ((y[3] / total) * 100) * 5]

sizes = porcentagem

colors = plt.get_cmap('Blues')(np.linspace(0.5, 1.1, len(x)))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.75)

plt.show()
```

## Gráfico em Pizza comparando o aumento entre os anos. ↓↓↓

```python
fig, ax = plt.subplots()

ano = ['2019', '2020', '2021', '2022']
valores = [pais['2019'].values[0], pais['2020'].values[0], pais['2021'].values[0], pais['2022'].values[0]]

colors = plt.get_cmap('Blues')(np.linspace(0.3, 0.9, len(valores)))

ax.pie(valores, labels=ano, autopct='%1.1f%%', colors=colors)

plt.show()
```
